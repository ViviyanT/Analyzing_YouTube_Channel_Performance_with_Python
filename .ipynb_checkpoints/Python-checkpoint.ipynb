## 2. Data Collection

Before we dive into the technical details, let's first gain an understanding of the data source. The data for this analysis was collected using the `YouTube API`, a powerful tool that enables us to retrieve channel statistics and video information directly from YouTube's platform.

### 2.1. Data Collection Process

The necessary data for analysis was obtained through the following key steps:

- **API Key Setup:** An API key was obtained from YouTube's developer console to authenticate and access the API's functionalities.
- **Building the API Client:** Using the `googleapiclient` library, a Python client was constructed to enable requests to the YouTube API.
- **Retrieving Channel Statistics:** Channel statistics, including subscriber count, view count, and video count, were requested and obtained.
- **Gathering Video Details:** Detailed information about individual videos, such as view count, like count, comment count, and more, was collected through API calls.

```python
from googleapiclient.discovery import build
from dateutil import parser
import pytz
import pandas as pd
from datetime import datetime
import os

# Data visualisation packages
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# NLP
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
from wordcloud import WordCloud
```

```python
# API Key Setup
api_service_name = "youtube"
api_version = "v3"
api_key = os.environ.get('YOUTUBE_API_KEY')

# Building the API Client
youtube = build(api_service_name, api_version, developerKey=api_key)
```

```python
# Retrieving Channel Statistics
channel_ids = ["UCSHZKyawb77ixDdsGog4iWA"]
              

def get_channel_stats(youtube, channel_ids): 
    
    """
    Retrieves channel statistics from the YouTube API.

        Args:
            youtube (googleapiclient.discovery.Resource): The YouTube API client.
            channel_ids (list): List of channel IDs to retrieve statistics for.

        Returns:
            pandas.DataFrame: DataFrame containing channel statistics.
    """
    
    all_data = []

    #API Request and Data Retrieval
    try:
        request = youtube.channels().list(
            part="snippet,contentDetails,statistics",
            id=",".join(channel_ids)
        )
        response = request.execute()

        # Processing API Response -> loops through items in the API response
        for item in response['items']:
            # Extract relevant data from the API response
            data = {
                'channelName': item['snippet']['title'],
                'subscribers': item['statistics']['subscriberCount'],
                'views': item['statistics']['viewCount'],
                'totalVideos': item['statistics']['videoCount'],
                'playlistId': item['contentDetails']['relatedPlaylists']['uploads']
            }
            all_data.append(data)

        return pd.DataFrame(all_data)

    except Exception as e:
        print("An error occurred:", e)
        return pd.DataFrame()  # Return an empty DataFrame if there was an error
    
channel_stats = get_channel_stats(youtube, channel_ids)
channel_stats
```

### Gathering Video Details

`playlistId` is a unique identifier, which represents the playlist containing a channel's uploaded videos. We will use it with the `playlistItems.list` method to retrieve a list of video IDs associated with the channel's uploads. Finally, we will use these IDs to retrieve detailed information about each individual video using the YouTube API.

```python
playlist_id = "UUSHZKyawb77ixDdsGog4iWA"

def get_video_ids(youtube, playlist_id):
    
    """
    Retrieves video IDs from a YouTube playlist.

    Args:
        youtube (googleapiclient.discovery.Resource): The YouTube API client.
        playlist_id (str): The ID of the playlist to retrieve video IDs from.

    Returns:
        list: List of video IDs.
    """
                                        
    video_ids = []

    # Initial request to retrieve video IDs
    request = youtube.playlistItems().list(
        part="contentDetails",
        playlistId=playlist_id,
        maxResults = 50
    )
    response = request.execute()
        
    # Loop through the initial response and collect video IDs
    for item in response['items']:
        video_ids.append(item['contentDetails']['videoId'])

    # Check if there are more pages with videos and retrieve their IDs
    next_page_token = response.get('nextPageToken')
    while next_page_token is not None:
        request = youtube.playlistItems().list(
            part='contentDetails',
            playlistId = playlist_id,
            maxResults = 50,
            pageToken = next_page_token
        )
        response = request.execute()

        # Loop through the response and collect video IDs
        for item in response['items']:
            video_ids.append(item['contentDetails']['videoId'])
                
        # Update the value of next_page_token for the next request
        next_page_token = response.get('nextPageToken')

    return video_ids

video_ids = get_video_ids(youtube, playlist_id)
len(video_ids)
```

```python
def get_video_details(youtube, video_ids):
    
    """
    Retrieves detailed information about YouTube videos.

    Args:
        youtube (googleapiclient.discovery.Resource): The YouTube API client.
        video_ids (list): List of video IDs to retrieve details for.

    Returns:
        pandas.DataFrame: DataFrame containing video details.
    """
        
    all_video_info = []
    
    # Process video IDs in batches of 50
    for i in range(0, len(video_ids), 50):
        # Request to retrieve video details for the batch
        request = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            id=','.join(video_ids[i:i+50])
        )
        response = request.execute() 

        # Loop through the response and collect video details
        for video in response['items']:
            stats_to_keep = {'snippet': ['channelTitle', 'title', 'description', 'publishedAt','tags'],
                             'statistics': ['viewCount', 'likeCount', 'favouriteCount', 'commentCount'],
                             'contentDetails': ['duration']
                            }
            
            video_info = {'video_id': video['id']}
            
            # Iterate through 'stats_to_keep' keys and values
            for k in stats_to_keep.keys():
                for v in stats_to_keep[k]:
                    try:
                        video_info[v] = video[k][v]
                    except:
                        video_info[v] = None

            all_video_info.append(video_info)
    
    
    return pd.DataFrame(all_video_info)


video_df = get_video_details(youtube, video_ids)
video_df
```

## 3. Data Cleaning and Preparation

With our dataset in hand, the next crucial step is to prepare the data for analysis. This involves cleaning, formatting, and structuring the data to ensure its quality and consistency. Let's explore how this important phase was tackled.

```python
# Check for NULL values
video_df.isnull().any()
```

Tags contain some NULL values, while 'favouriteCount' has only NULL values. To ensure data quality, the `favouriteCount` column was removed, and further investigation is needed for the `tags` column.

```python
# Delete unused NULL values
video_df = video_df.drop(columns = ['favouriteCount'])
```

```python
# Add tag count
video_df['tagCount'] = video_df['tags'].apply(lambda x: 0 if x is None else len(x))
```

```python
# Check data types
video_df.dtypes
```

As we can see several columns containing counts are not of numeric data type. To facilitate analysis, let's convert the following columns to numeric data types:

```python
# Convert count columns to numeric data types
numeric_cols = ['viewCount', 'likeCount', 'commentCount']
video_df[numeric_cols] = video_df[numeric_cols].apply(pd.to_numeric, errors = 'coerce', axis = 1)
```

In order to perform some time-related analysis, let's convert the `publishedAt` column to a datetime format and extract from it which day of the week the video was uploaded on.

```python
# Convert 'publishedAt' in datetime format
video_df['publishedAt'] = pd.to_datetime(video_df['publishedAt'])

# Extract the day of the week and store it in a new column
video_df['publishDayName'] = video_df['publishedAt'].dt.strftime("%A")
```

Finally, the duration needs to be converted into something easier to understand, like minutes. And because we can, let's extract the video guest names from `title`.

```python
# Convert 'duration' to minutes
import isodate
video_df['durationMins'] = video_df['duration'].apply(lambda x: isodate.parse_duration(x))
video_df['durationMins'] = video_df['durationMins'].astype('timedelta64[m]')
video_df['durationMins'] = video_df['durationMins'].astype(int)
```

```python
import re
# Extract the Podcast Guest from 'title'
def extract_guest_name(title):
    match = re.match(r'^([^:]+):', title)
    if match:
        return match.group(1).strip()  # Remove spaces
    else:
        return None

# Apply the function to create a new 'guest' column
video_df['guest'] = video_df['title'].apply(extract_guest_name)
```

After applying these data preparation steps, our dataset is now cleaned, formatted, and ready for in-depth analysis.

## 4. Exploratory Data Analysis (EDA) and Visualizations

Let's begin this analysis by looking at the heatmap below. It illustrates the correlations between numeric columns, shedding light on the relationships between various engagement metrics. 

### Heatmap

```python
plt.figure(figsize=(10, 5))
heatmap = sns.heatmap(video_df.corr(numeric_only=True), annot=True, annot_kws={"size": 12})

heatmap.set_xticklabels(heatmap.get_xticklabels(), fontsize=12)
heatmap.set_yticklabels(heatmap.get_yticklabels(), fontsize=12)

plt.show()
```

As expected, we observe the strongest correlations between `view count` and metrics like `likes` and `comments`. Higher view counts tend to coincide with increased likes and comments. Additionally, there is a 0.42 correlation between video duration (in minutes) and the views count, which might be worth exploring.

'tagCount' exhibits minimal correlation with the engagement metrics and won't be further analyzed.

Now, let's explore the top 3 metrics with the highest correlation to `viewCount` - `likeCount`, `commentCount`, and `durationMins`!

### Views vs. likes and comments

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Scatter plot: View Count vs. Likes Count
sns.scatterplot(data=video_df, x='likeCount', y='viewCount', ax=ax[0])
ax[0].set_title('View Count vs. Likes Count')

# Scatter plot: Views Count vs. Comments Count
sns.scatterplot(data=video_df, x='commentCount', y='viewCount', ax=ax[1])
ax[1].set_title('Views Count vs. Comments Count')

# Set tick label font size for both subplots
for axis in ax:
    axis.tick_params(axis='both', which='major', labelsize=10)

# Adjust spacing between subplots
plt.tight_layout(pad=2.0)

# Show the plots
plt.show()
```

The scatter plots highlight the strongest correlations. Notably, higher `likeCount` generally coincides with increased `viewCount`. In contrast, while a moderate `commentCount` correlates with higher views, the relationship becomes less consistent beyond a certain point. These findings highlight the close relationship between likes, comments, and video views.

Next, let's examine `video duration`, a particularly noteworthy aspect of Lex Fridman's channel due to its podcast-like content. The heatmap suggested that duration might have an impact on engagement metrics, and now we'll explore this relationship with a focus on views, likes, and comments.

### Duration and the metrics

```python
duration_analysis = video_df.groupby('durationMins')[['viewCount']].mean()

plt.figure(figsize=(10, 6))

plt.plot(duration_analysis.index, duration_analysis['viewCount'], label='Avg View Count')

plt.title('Video Duration Analysis: Average Views vs. Duration')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Average Views')
plt.legend()

plt.gca().get_yaxis().set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.grid(True)
plt.show()
```

In our video duration analysis, we observed interesting trends. Videos within the 0-100 minute range start with around 2.5 million views, and from 100-200 minutes, views increase significantly to nearly 6.5 million. A remarkable peak of approximately 10.5 million views is noted around 240 minutes, possibly driven by specific high-engagement videos within that duration range.

Longer videos, while not reaching the same peak, still maintain a consistent level of views. 

Lex Fridman's videos tend to have great view counts across various durations. However, there are specific durations, particularly in the `100 to 200-minute range`, that tend to attract exceptionally high engagement. 

Let's perform this same analysis on the likes and comments and see whether we will find something interesting.

```python
duration_analysis = video_df.groupby('durationMins')[['likeCount', 'commentCount']].mean()

plt.figure(figsize=(10, 6))

plt.plot(duration_analysis.index, duration_analysis['likeCount'], label='Avg Like Count')
plt.plot(duration_analysis.index, duration_analysis['commentCount'], label='Avg Comment Count')

plt.title('Video Duration Analysis: Average Likes and Comments vs. Duration')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Average Likes and Comments')
plt.legend()

plt.gca().get_yaxis().set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.grid(True)
plt.show()
```

We can see a similar trend in the relationship between video duration and the number of likes and comments. 

Videos in the `100-200-minute range` consistently receive the highest number of likes and comments, with a notable peak in likes around the 240-minute mark, followed by a relatively steady level of engagement.

Now that we've concluded our examination of the numerical data, let's shift our focus to exploring how categorical data, such as `publishDayName` and later `guest`, can provide further insights into this channel's performance.

### Does the day of the week a video is published influence its engagement?

```python
# Define weekday_order
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Calculate video counts by publishDayName
video_counts_by_day = video_df['publishDayName'].value_counts().reindex(weekday_order, fill_value=0)

# Create a bar plot for video posting frequency
plt.figure(figsize=(10, 6))
sns.barplot(x=video_counts_by_day.index, y=video_counts_by_day.values, palette='viridis')
plt.title('Video Posting Frequency by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Videos Posted')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Group by publishDayName and calculate mean engagement metrics
engagement_by_day = video_df.groupby('publishDayName')[['viewCount', 'likeCount', 'commentCount']].mean()
engagement_by_day = engagement_by_day.reindex(weekday_order)

# Plotting average engagement metrics by day
plt.figure(figsize=(10, 6))
ax = engagement_by_day.plot(kind='bar', colormap='viridis')
plt.title('Average Engagement Metrics by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Count')
plt.xticks(rotation=45)
plt.legend(['Views', 'Likes', 'Comments'])
plt.tight_layout()

plt.show()
```

It's worth noting that Lex Fridman's upload schedule appears to roughly align with his analytics of average views for each day of the week. While there may be room for optimization, such as **posting more frequently on Tuesdays instead of Mondays**, the **influence on real-life engagement seems minimal**. Interestingly, Wednesday and Sunday appear to be the days with lower posting frequency and lower engagement levels.

Now, let's shift our focus to examining which guest appearances have collected the best engagement metrics on Lex Fridman's channel.

### Guest Impact on Engagement

```python
guest_engagement = video_df.groupby('guest')[['viewCount', 'likeCount', 'commentCount']].mean()

# Sort the guests by average view count in descending order
guest_engagement = guest_engagement.sort_values('viewCount', ascending=False)

# Plotting the results
plt.figure(figsize=(12, 6))
sns.barplot(data=guest_engagement.head(8), x='viewCount', y=guest_engagement.index[:8])
plt.xlabel('Average View Count')
plt.ylabel('Podcast Guest')
plt.title('Top 8 Podcast Guests by Average View Count')
plt.show()
```

```python
def top_n_frequent_guests(video_df, n=1):
    # Filter videos where there is a guest (guest is not None)
    videos_with_guest = video_df[video_df['guest'].notna()]
    
    # Count the number of appearances for each guest
    guest_counts = videos_with_guest['guest'].value_counts()
    
    # Select the top N most frequent guests
    top_n_guests = guest_counts.head(n)
    
    # Create a bar plot to visualize the top N most frequent guests
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_n_guests.index, y=top_n_guests.values, palette='viridis')
    plt.title(f'Top {n} Most Frequent Guests')
    plt.xlabel('Guest Names')
    plt.ylabel('Number of Appearances')
    plt.yticks([])
    plt.tight_layout()
    plt.show()
    
    # Return the top N most frequent guests and their appearance counts
    return top_n_guests

top_n_guests = top_n_frequent_guests(video_df, n=5)
```

Surprisingly, the guests with the highest average view counts don't necessarily align with the most frequently appearing guests on Lex Fridman's channel. This intriguing discovery suggests that **audience engagement may depend on a variety of factors beyond guest popularity**.

From this time series analysis, we can observe that Lex's engagement was steadily growing until the middle/end of 2022, after which it began to decline. 

To better understand this trend, let's visualize Lex's video posting schedule during the months from mid-2022 to the present.

```python
video_df['publishedAt'] = pd.to_datetime(video_df['publishedAt'], utc=True)

# Filter data for dates from mid-2022 onwards
start_date = pd.Timestamp(year=2022, month=6, day=1, tz='UTC')
filtered_df = video_df[video_df['publishedAt'] >= start_date]

# Create a new DataFrame with 1 for days with videos and NaN for days without videos
video_schedule = pd.DataFrame({'Video Posted': 1}, index=filtered_df['publishedAt'].dt.date.unique())
video_schedule.index = pd.to_datetime(video_schedule.index)  # Ensure the index is in DateTime format

# Create a date range without specifying the timezone
date_range = pd.date_range(start=start_date.date(), end=pd.Timestamp.now(tz='UTC').date())

# Create a new DataFrame filled with NaN for all dates in the range
full_video_schedule = pd.DataFrame({'Video Posted': None}, index=date_range)

# Update the full schedule with ones for days with video posts
full_video_schedule.update(video_schedule)

plt.figure(figsize=(12, 6))

plt.plot(full_video_schedule.dropna().index, [1] * len(full_video_schedule.dropna()), marker='o', linestyle='', color='b', markersize=8)

plt.title('Video Posting Schedule (Mid-2022 Onwards)')
plt.xlabel('Date')
plt.ylabel('Video Posted')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

It's worth noting that Lex Fridman's posting frequency decreased from January 2023 onwards, possibly due to his travels to Ukraine.

As a result, the total number of videos posted decreased during this period, which naturally led to lower overall engagement for the channel. In essence, the explanation for the observed decline in engagement metrics is that **Lex has been posting less frequently**.
